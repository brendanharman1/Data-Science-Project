First we must import the pandas package and read in the ranking data.

```
import pandas as pd
df_rank = pd.read_csv("player_ranking_2024.csv")
df_rank["AVERAGE POINTS"] = df_rank["AVERAGE POINTS"].str.replace("-", "0").astype(float)
df_rank["TOTAL POINTS"] = df_rank["TOTAL POINTS"].str.replace("-","0").astype(float)
df_rank["NAME"] = df_rank["NAME"].str.replace('Cameron Davis', 'Cam Davis')
df_rank["NAME"] = df_rank["NAME"].str.replace('Ludvig Aberg', 'Ludvig Åberg')
df_rank
```

The output is a cleaned dataset that shows a player, and their various statistics related to ranking. We will be focused on the END column as that shows the end of 2024 ranking.

Next we must read in the statistical data which is already cleaned. 

```
df_stats = pd.read_csv("player_stats_2024.csv")
df_stats
```

The output shows us a dataset with players and their various statistics from the 2024 season as captured by ESPN. 

Next we must merge the two data sets into one, merging on name, so that we can then continue our analysis using the merged data. 

```
df_merge = df_stats.merge(df_rank, left_on="Name", right_on="NAME", how='left')
df_merge = df_merge[['Name','AGE', 'CUP','EVNTS', 'RNDS', 'CUTS', 'TOP10', 'WINS','SCORE', 'DDIS', 'DACC', 'GIR', 'PUTTS', 'SAND', 'BIRDS', 'END', 'AVERAGE POINTS', 'TOTAL POINTS']]
df_merge
```

The output shows the merged dataset with all players from the stats dataset and their related ranking. This also only includes relevant columns to analysis and cleans up redundant/unnecessary data.

In the next two cells we need to check that the merge was successful and outputs what we would like.

```
df_stats.shape[0] == df_merge.shape[0]
```

This code shows that on the merge we get the expected number of row since we performed a left merge on df_stats. This means that there was no extra data added.

```
df_merge.isnull().any().any()
```

This shows us there is no nan/null data and all of the stats rows are populated with ranking data. 

We would like to now run statistical analysis to find which statistical metrics are correlated to the world rankings of a player. To do this we will use Spearman Rank Order Correlation since we wish to compare ranked data.

```
from scipy.stats import spearmanr

#split data into rank and stat columns
ranking_col = df_merge["END"]
statistical_cols = df_merge[["SCORE", "DDIS", "DACC", "GIR", "PUTTS", "SAND", "BIRDS"]]

results = {}

#run the analysis on each column and store the results in a dictionary
for col in statistical_cols.columns:

    stat_col = statistical_cols[col]

    rho, p_value = spearmanr(ranking_col, stat_col)
    
    results[col] = {
        'Spearman_Rho': rho,
        'P_Value': p_value
    }

results_df = pd.DataFrame(results).T
results_df
```

The output shows the correlation value in one column and the p-value measuring significance in another column. From this analysis we can see that Score, DDIS, adn GIR are significant based on a 95% confidence interval.
